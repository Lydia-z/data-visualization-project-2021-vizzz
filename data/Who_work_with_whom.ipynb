```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./full_data.csv', error_bad_lines=False)
print('size of the third dataset is ', data.shape)
data.head()
```

```python
directors ={} #{name: count}
producers = {}
music_directors = {}

directors_edges = [] #[(source, target)...]
producers_edges = []
music_directors_edges = []
```

```python
def get_nodes_edges(string, nodes, links):
    if(string == 'nan'):
        return nodes, links
    # if the string is with []
    #print(string)
    if('[' in string and ']' in string):
        # get rid of the []
        string = string[1:-1]
        
    names = sorted(string.split(', '))
    for name in names:
        # if name string with ''
        if('\'' in name):
            name = name[1:-1]
        # get rid of '(' or ')' or '(blahblahblah)' as they are all explaination
        if(name.startswith('(')):
            continue
        if(name.startswith(')')):
            continue
        # get rid of animation: ... also explaination
        if(name[-1]==':'):
            continue
        # get rid of xx director
        if('director' in name):
            continue
        # get rid of see program
        if('program' in name):
            continue
            
        if name in nodes:
            nodes[name] += 1
        else:
            nodes[name] = 1
    
    for i in range(len(names)):
        # if name string with ''
        if('\'' in names[i]):
            names[i] = names[i][1:-1]
        # get rid of '(' or ')' or '(blahblahblah)' as they are all explaination
        if(names[i].startswith('(')):
            continue
        if(names[i].startswith(')')):
            continue
        if(names[i][-1]==':'):
            continue
        # get rid of xx director
        if("director" in names[i]):
            continue
        if("program" in names[i]):
            continue
            
        for j in range(i+1, len(names)):
            # if name string with ''
            if('\'' in names[j]):
                names[j] = names[j][1:-1]
                # get rid of '(' or ')' or '(blahblahblah)' as they are all explaination
            if(names[j].startswith('(')):
                continue
            if(names[j].startswith(')')):
                continue
            if(names[j][-1]==':'):
                continue
            if("director" in names[j]):
                continue
            if("program" in names[j]):
                continue
            links.append((names[i],names[j]))
            
    return nodes, links
```

```python
for ind, row in data.iterrows():
    if(pd.notna(row['Directed by'])):
        directors,directors_edges = get_nodes_edges(row['Directed by'],directors,directors_edges)
    if(pd.notna(row['Produced by'])):
        producers,producers_edges = get_nodes_edges(row['Produced by'],producers,producers_edges)
    if(pd.notna(row['Music by'])):
        music_directors,music_directors_edges = get_nodes_edges(row['Music by'],music_directors,music_directors_edges)
```

```python
import collections
counter=collections.Counter(directors_edges)
```

```python
import json

directors = {"nodes": director_nodes, "links":director_links}

with open('directors.json', 'w') as json_file:
  json.dump(directors, json_file)
```

```python
producers_nodes = [{"name":key,"count":producers[key]} for key in producers]
counter=collections.Counter(producers_edges)
producers_links = [{"source":key[0],"target":key[1],"weight":counter[key]} for key in counter]
```

```python
producers = {"nodes": producers_nodes, "links":producers_links}

with open('producers.json', 'w') as json_file:
  json.dump(producers, json_file)
```

```python
music_directors_nodes = [{"name":key,"count":music_directors[key]} for key in music_directors]
counter=collections.Counter(music_directors_edges)
music_directors_links = [{"source":key[0],"target":key[1],"weight":counter[key]} for key in counter]

music_directors = {"nodes": music_directors_nodes, "links":music_directors_links}

with open('music_directors.json', 'w') as json_file:
  json.dump(music_directors, json_file)
```
